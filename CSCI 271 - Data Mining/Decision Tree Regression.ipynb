## Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor as DT
from sklearn.tree import plot_tree
```

## Dataset

```python
df = pd.read_csv('Decision Tree Regression Report - Dataset.csv')
df
```

## Splitting

```python
X_train = df.iloc[:10,:-1]
X_train
```

```python
X_test = df.iloc[-2:,:-1]
X_test
```

```python
y_train = df.iloc[:10,-1]
y_train
```

## Modeling

```python
dt_regressor = DT(random_state=36)
dt_regressor.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt_regressor, feature_names=list(df.columns))
plt.show()
```

```python
dt_regressor.predict(X_test)
```

```python
dt_regressor_2 = DT(random_state=36, criterion='absolute_error')
dt_regressor_2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt_regressor_2, feature_names=list(df.columns))
plt.show()
```

```python
dt_regressor_2.predict(X_test)
```
