```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, MiniBatchKMeans
```

```python
column_names = ["X1", "X2"]  
df = pd.read_csv("2d-data.csv", header=None, names=column_names)
```

```python
df.head()
```

```python
def plot_clusters(df, k):
    kmeans = KMeans(n_clusters=k, random_state=42)
    df["Cluster"] = kmeans.fit_predict(df)

    plt.figure(figsize=(6, 4))
    for cluster in range(k):
        cluster_points = df[df["Cluster"] == cluster]
        plt.scatter(cluster_points["X1"], cluster_points["X2"], label=f"Cluster {cluster}")

    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                c="red", marker="x", s=200, label="Centroids")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.title(f"K-Means Clustering (k={k})")
    plt.legend()
    plt.show()

# Generate and plot clusters for k = 3 and k = 4
plot_clusters(df.copy(), k=3)
plot_clusters(df.copy(), k=4)
```

```python
k_values = range(2, 11)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df)
    inertia_values.append(kmeans.inertia_)

# Plot inertia vs. number of clusters
plt.figure(figsize=(6, 4))
plt.plot(k_values, inertia_values, marker="o", linestyle="-", color="b")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

Based on the elbow method, k = 3 is the optimal number of clusters.

```python
def compare_kmeans_methods(random_state):
    k = 4  # Number of clusters
    
    # K-Means clustering
    kmeans = KMeans(n_clusters=k, random_state=random_state, n_init=1)
    kmeans.fit(df)
    
    # Mini-Batch K-Means clustering
    mini_kmeans = MiniBatchKMeans(n_clusters=k, random_state=random_state, n_init=1, batch_size=50)
    mini_kmeans.fit(df)
    
    print(f"Random State: {random_state}")
    print(f"K-Means -> Iterations: {kmeans.n_iter_}, Inertia: {kmeans.inertia_:.2f}")
    print(f"Mini-Batch K-Means -> Iterations: {mini_kmeans.n_iter_}, Inertia: {mini_kmeans.inertia_:.2f}")
    print("-" * 50)

    # Plot clusters
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # K-Means Clusters
    df["Cluster"] = kmeans.labels_
    axes[0].scatter(df["X1"], df["X2"], c=df["Cluster"], cmap="viridis", alpha=0.6)
    axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
                    c="red", marker="x", s=200, label="Centroids")
    axes[0].set_title("K-Means Clustering")
    
    # Mini-Batch K-Means Clusters
    df["Mini_Cluster"] = mini_kmeans.labels_
    axes[1].scatter(df["X1"], df["X2"], c=df["Mini_Cluster"], cmap="viridis", alpha=0.6)
    axes[1].scatter(mini_kmeans.cluster_centers_[:, 0], mini_kmeans.cluster_centers_[:, 1], 
                    c="red", marker="x", s=200, label="Centroids")
    axes[1].set_title("Mini-Batch K-Means Clustering")

    for ax in axes:
        ax.set_xlabel("X1")
        ax.set_ylabel("X2")
        ax.legend()
    
    plt.show()

random_states = [41, 99, 123] 
for rs in random_states:
    compare_kmeans_methods(rs)
```

Among the three random states tested, the inertia for the mini batch K Means was higher than the inertia for the regular K Means. When looking at the graph of the clusters, the mini batch K Means produced more heterogeneous clusters of data points compared to the regular K Means. In the mini batch K Means graph, the centroids of two of the clusters are very close in proximity, which leads to unclear boundaries for the respective clusters. 

Since the mini batch K Means algorithm only takes a sample of the entire dataset when training, the convergence time in finding a solution is reduced. This can be seen in the number of iterations where the mini batch K Means only went through one iteration of the entire dataset.
