```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, SpectralClustering, MeanShift, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("lab7_dataset.csv", header = None)
data = data.rename(columns = {0: 'X1',1: 'X2'})
data.head()
```

```python
models = {
    'KMeans': KMeans(n_clusters = 3),
    'Gaussian Mixture': GaussianMixture(n_components = 3),
    'Ward': AgglomerativeClustering(n_clusters = 3, linkage = 'ward'),
    'Spectral': SpectralClustering(n_clusters = 3, affinity = 'nearest_neighbors'),
    'MeanShift': MeanShift()
}
```

```python
cluster_results = {}
for name, model in models.items():
    if isinstance(model, GaussianMixture):
        cluster_results[name] = model.fit_predict(data)
    else:
        cluster_results[name] = model.fit(data).labels_
```

```python
# The following code was obtained from ChatGPT. The purpose is to solely reassign the cluster assignments using the KMeans labels
# so that the colors and clusters assignments stay consistent across models. This is primarily made for ease in visual comparisons
# for the six points between different models.

# Use KMeans as the reference clustering
reference_labels = cluster_results['KMeans']

def align_labels(ref_labels, pred_labels):
    """Reassign cluster labels to match reference labels as closely as possible."""
    cm = confusion_matrix(ref_labels, pred_labels)
    row_ind, col_ind = linear_sum_assignment(-cm)  # Maximize agreement
    mapping = {col: row for row, col in zip(row_ind, col_ind)}
    return np.array([mapping[label] for label in pred_labels])

for name in models.keys():
    if name != 'KMeans':  # Skip reference
        cluster_results[name] = align_labels(reference_labels, cluster_results[name])
```

```python
fig, axes = plt.subplots(2, 3, figsize = (12, 8))
axes = axes.flatten()

for ax, (name, labels) in zip(axes, cluster_results.items()):
    ax.scatter(data['X1'], data['X2'], c = labels, edgecolor = 'k')
    ax.set_title(name)

plt.tight_layout()
plt.show()

key_points = [(-15, 5), (10, 9)] + list(data.nsmallest(4, 'X2')[['X1', 'X2']].itertuples(index = False, name = None))

key_memberships = {}
for name, labels in cluster_results.items():
    key_memberships[name] = [int(labels[np.argmin(np.linalg.norm(data.values - np.array(p), axis = 1))]) for p in key_points]

print("Cluster Assignments for each point:")
for method, memberships in key_memberships.items():
    print(f"{method}: {memberships}")
```

# Discussion
### Gaussian Mixture: 
We visually observe that the clusters vary in densities. GMM assumes that data is generated from multiple Gaussian distributions and assigns points probabilistically. Compared to K-means algorithm, the northern compact cluster is considered its own, and does not include the "isolated data points" around it. The same property goes for the eastern cluster that is less compact. Therefore, the more scattered data points (including the top left and top right points) are assigned to the southwest cluster. All 6 points belong to the southwest cluster.

### Ward
This method relies on iteratively merging the two closest clusters in a way that minimizes the increase in total variance. As an effect, we observe less spherical clustering (more espectially evident for the north cluster) compared to the clusters produced by the KMeans algorithm. Consequently, the top left point belongs to the southwest cluster. The top right, similar to its KMeans counterpart, belongs to the northeast cluster. The 4 points (of y-values less than -17) belong to the southwest cluster.

### Spectral
Speaking with respect to the geometric nature of clusters produced in spectral clustering, the algorithm attempts to capture non-linear data to produce complex cluster structures. However, given this specific 2-D data, the effects of the clustering is the same to KMeans for the six points. Though it may be observed that the clusters in Spectral algorithm looks empirically similar to the clusters produced by the Ward algorithm.

### Mean Shift:
Mean Shift tries to find areas on the graph with highest density. In this case, it is effective for the north and east clusters. Because of the sparsity of the lower left cluster, its effect results in 4 clusters. The top left belong to its own new cluster. The top right belong to the same cluster it belongs to in KMeans. Aside from this this, the other clusters have similar characteristics with the clusters produced via the KMeans algorithm. The 4 points (of y-values less than -17) belong to the southwest cluster.
