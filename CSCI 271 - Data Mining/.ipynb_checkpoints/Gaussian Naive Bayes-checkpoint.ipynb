## Scikit-learn

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

# Train the model on the dataset.
gnb.fit(X,y)

```

```python
# Test instances which the Gaussian NB model will classify
x_1 = [[5.0, 3.5, 1.3, 0.3]]
x_2 = [[5.5, 2.6, 4.4, 1.2]]
x_3 = [[6.7, 3.1, 5.6, 2.4]]

prediction_x1 = gnb.predict(x_1)
probability_x1 = gnb.predict_proba(x_1)
```

```python
# Find the Mean and Standard Deviation

summary_setosa = combined[combined["class"] == "Iris-setosa"][["sepal length", "sepal width", "petal length", "petal width"]].agg(['mean', lambda x: x.std(ddof=0)]).rename(index={'<lambda>': 'std'})
summary_versicolor = combined[combined["class"] == "Iris-versicolor"][["sepal length", "sepal width", "petal length", "petal width"]].agg(['mean', lambda x: x.std(ddof=0)]).rename(index={'<lambda>': 'std'})
summary_virginica = combined[combined["class"] == "Iris-virginica"][["sepal length", "sepal width", "petal length", "petal width"]].agg(['mean', lambda x: x.std(ddof=0)]).rename(index={'<lambda>': 'std'})

```

```python
# Relative Likelihood

def likelihood(feature_value, mean, std):
    a = 1/(std*np.sqrt(2*np.pi))
    b = np.exp(-0.5 * ((feature_value - mean) / std)**2)
    return a*b

```

```python
# Classifying x_1

score_x1_setosa = prior_setosa * likelihood(5.0, 5.037500, 0.357552) * likelihood(3.5, 3.440000, 0.359722) * likelihood(1.3, 1.462500, 0.169834	) * likelihood(0.3, 0.232500, 0.098457)
score_x1_versicolor = prior_versicolor * likelihood(5.0, 6.010000, 0.516624) * likelihood(3.5, 2.780000	, 0.328786) * likelihood(1.3, 4.317500, 0.44547) * likelihood(0.3, 1.35000, 0.204939)
score_x1_virginica = prior_virginica * likelihood(5.0, 6.622500, 0.675458) * likelihood(3.5,2.960000, 0.332265	) * likelihood(1.3, 5.607500, 0.580253) * likelihood(0.3, 1.9900,0.269072)
```
