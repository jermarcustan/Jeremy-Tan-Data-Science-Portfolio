# Ward Linkage_Crime Data Example

## Data Prep

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
column_names = ["City","Murder","Rape","Robbery","Assault","Burglary","Larceny","AutoTheft"]

df = pd.read_table("CityCrime.dat", names=column_names, sep="\t")

crime_data = df[["Murder", "Rape", "Robbery", "Assault", "Burglary", "Larceny", "AutoTheft"]]

crime_data
```

## Merge Matrix

```python
crime_data.fillna(crime_data.mean(), inplace=True)

Z = linkage(crime_data, method="ward")

merge_matrix_df = pd.DataFrame(
    Z[:],
    columns=["Cluster1", "Cluster2", "Distance", "NewClusterSize"]
)

merge_matrix_df
```

## Dendogram

```python
plt.figure(figsize=(10, 5))

# Dendogram
dendrogram(Z, labels=df.index, leaf_rotation=90, leaf_font_size=8)

# Chart Labels
plt.title("Ward Linkage on Crime Data Set")
plt.xlabel("City")
plt.ylabel("Distance")
plt.show()
```

### t = 1000

```python
plt.figure(figsize=(10, 5))

# Dendrogram
dendrogram(Z, labels=crime_data.index, leaf_rotation=90, leaf_font_size=10)

# Threshold Line
plt.axhline(y=1000, color='r', linestyle='--', label="Cluster Threshold")
clusters = fcluster(Z, t=1000, criterion='distance') # Groups cities into clusters

# Chart Labels
plt.title("Ward Linkage on Crime Data Set (t=1000)")
plt.xlabel("City")
plt.ylabel("Distance")
plt.legend()
plt.show()

# Displays cities within clusters
crime_clusters_df = pd.DataFrame({'City': crime_data.index, 'Cluster': clusters})
crime_clusters_df.groupby("Cluster")["City"].apply(list)
```

### t = 500

```python
plt.figure(figsize=(10, 5))

# Dendrogram
dendrogram(Z, labels=crime_data.index, leaf_rotation=90, leaf_font_size=10)

# Threshold Line
plt.axhline(y=500, color='r', linestyle='--', label="Cluster Threshold")
clusters = fcluster(Z, t=500, criterion='distance') # Groups cities into clusters

# Chart Labels
plt.title("Ward Linkage on Crime Data Set (t=500)")
plt.xlabel("City")
plt.ylabel("Distance")
plt.legend()
plt.show()

# Displays cities within clusters
crime_clusters_df = pd.DataFrame({'City': crime_data.index, 'Cluster': clusters})
crime_clusters_df.groupby("Cluster")["City"].apply(list)
```
