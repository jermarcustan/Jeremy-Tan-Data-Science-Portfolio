```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def inverse_transform_sampling(n):
    """
    Generate random variates of the random variable X using the inverse transform method.
    
    Parameters:
        n (int): Number of random variates to generate.
    
    Returns:
        numpy.ndarray: Array of size n containing the generated random variates.
    """
    # Generate n uniform random variates on [0, 1]
    u = stats.uniform.rvs(size=n)
    
    # Compute random variates of X using inverse transform method
    x = np.sqrt(-np.log(1 - u))
    
    return x

# Example usage
n_samples = 10000
samples = inverse_transform_sampling(n_samples)

# Display generated samples
print("Generated Samples:")
print(samples)

plt.hist(samples, bins=20, density=True)
# Plots the function f(x) on the interval [0,3]
xs = np.linspace(0, 3, 1000) # generate 1000 evenly-spaced points over [0,3]
ys = 2*xs*np.exp(-(xs*xs))
plt.plot(xs, ys)
```

# 2

```python
import math
import numpy as np
import random as rand
import matplotlib.pyplot as plt

C=2

# target pdf: 
def f(x):
    return 2*x*math.exp(-x*x)

# proposal pdf: exp(1)
def g(x):
    return math.exp(-x)


num_accepted = 0
num_rejected = 0

def accept_reject():
    global num_accepted, num_rejected
    while True:
        x = rand.expovariate(1)
        y = rand.uniform(0, C*g(x))
        if y <= f(x):
            num_accepted += 1
            break
        else:
            num_rejected += 1
    return x

samples = [accept_reject() for _ in range(10000)]
print(f"accepted {num_accepted} points out of {num_accepted + num_rejected}")
print(f"observed empirical efficiency {num_accepted / (num_accepted + num_rejected)}") 

plt.hist(samples, bins = 50, density=True)
# Plots the function f(x) on the interval [0,5]
xs = np.linspace(0, 5, 1000) # generate 1000 evenly-spaced points over [0,5]
ys = 2*xs*np.exp(-xs*xs)
plt.plot(xs, ys)
```

```python
import random
import numpy as np

def roll_die():
    return random.randint(1, 6)

def play_game():
    """
    Rules:
    1. Each player starts OFF the board (start at 0).
    2. Players use a fair 6-sided die.
    3. When players land on the bottom of a ladder, they move their piece to the top of the ladder.
    4. When players land on the head of a snake, they move their piece to the tail of the snake.
    5. The first player to get to or past 20 wins.
    """
    position = 0
    tosses = 0
    while position < 20:
        tosses += 1
        dice_roll = roll_die()
        position += dice_roll
        # Apply snake or ladder if landed on one
        if position in ladder:
            position = ladder[position]
        elif position in snake:
            position = snake[position]
    return tosses

# Game setup
ladder = {3: 11, 15: 19}
snake = {13: 4, 17: 10}

# Number of simulation runs
N = 1000

# Simulate the game N times
results = [play_game() for _ in range(N)]

# Calculate the average number of tosses required
mean_tosses = np.mean(results)

# Calculate the standard error
std_error = np.std(results) / np.sqrt(N)

# Calculate the 95% confidence interval
margin_of_error = 1.96 * std_error
lower_bound = mean_tosses - margin_of_error
upper_bound = mean_tosses + margin_of_error

print(f"Estimated Expected Value (E[X]): {mean_tosses:.3f}")
print(f"95% Confidence Interval: [{lower_bound:.3f}, {upper_bound:.3f}]")
```

```python
import random as rand
import matplotlib.pyplot as plt

def simulate(initial_state, end_time):
    states = [initial_state] # the jump chain y0,y1,y2,...
    times = [0] # the jump times j1,j2,j3,...
    
    while times[-1] < end_time: # while current time < end time
        # retrieve current state of simulation
        curr_state = states[-1] # y_n
        curr_time = times[-1] # j_n
        # compute next state and time
        if curr_state == 1:
            next_state = rand.choices([2,3], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
        elif curr_state == 2:
            next_state = 1
            holding_time = rand.expovariate(1/4)
        elif curr_state == 3:
            next_state = rand.choices([1,4], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
        elif curr_state == 4:
            next_state = rand.choices([3,5,6], weights=[1,1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
        elif curr_state == 5:
            next_state = rand.choices([4,6], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/4)
        else: # curr_state == 6
            next_state = rand.choices([4,5], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/4)
            
        # update simulation
        curr_time += holding_time
        
        # If the next time exceeds the end time, adjust it to end_time
        if curr_time > end_time:
            curr_time = end_time
        
        times.append(curr_time)
        states.append(next_state)
        
    return states, times

end_time = 10800
Y, J = simulate(1, end_time)
plt.scatter(J, Y)
for i in range(len(J)-1):
    plt.hlines(Y[i], J[i], J[i+1])

# Adjust y-axis ticks and labels
plt.yticks(range(1, 7), ['1', '2', '3', '4', '5', '6'])

plt.gca().set_aspect(1080)
plt.show()
```

```python
import random as rand
import numpy as np
from scipy.stats import norm

def simulate_with_total_time(initial_state, end_time):
    states = [initial_state] # the jump chain y0,y1,y2,...
    times = [0] # the jump times j1,j2,j3,...
    
    # Initialize total time spent in each room
    total_times = [0] * 6
    
    while times[-1] < end_time: # while current time < end time
        # retrieve current state of simulation
        curr_state = states[-1] # y_n
        curr_time = times[-1] # j_n
        
        # compute next state and time
        if curr_state == 1:
            next_state = rand.choices([2,3], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
            total_times[0] += holding_time
        elif curr_state == 2:
            next_state = 1
            holding_time = rand.expovariate(1/4)
            total_times[1] += holding_time
        elif curr_state == 3:
            next_state = rand.choices([1,4], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
            total_times[2] += holding_time
        elif curr_state == 4:
            next_state = rand.choices([3,5,6], weights=[1,1,1], k=1)[0]
            holding_time = rand.expovariate(1/2)
            total_times[3] += holding_time
        elif curr_state == 5:
            next_state = rand.choices([4,6], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/4)
            total_times[4] += holding_time
        else: # curr_state == 6
            next_state = rand.choices([4,5], weights=[1,1], k=1)[0]
            holding_time = rand.expovariate(1/4)
            total_times[5] += holding_time
            
        # update simulation
        curr_time += holding_time
        
        # If the next time exceeds the end time, adjust it to end_time
        if curr_time > end_time:
            curr_time = end_time
        
        times.append(curr_time)
        states.append(next_state)
        
    return total_times

# Number of simulation runs
N = 1000

# End time for each simulation run
end_time = 10800

# List to store total times for each room for each simulation run
all_total_times = []

# Run simulations
for _ in range(N):
    total_times = simulate_with_total_time(1, end_time)
    all_total_times.append(total_times)

# Convert to numpy array for easy calculations
all_total_times = np.array(all_total_times)

# Calculate sample mean and standard deviation for each room
means = np.mean(all_total_times, axis=0)
std_devs = np.std(all_total_times, axis=0)

# Calculate 95% confidence intervals for each room
z_score = norm.ppf(0.975) # 95% confidence level
lower_bound = means - z_score * (std_devs / np.sqrt(N))
upper_bound = means + z_score * (std_devs / np.sqrt(N))

# Calculate relative error for each confidence interval
relative_errors = std_devs / (means * np.sqrt(N))

# Print results
room_labels = ['Room 1', 'Room 2', 'Room 3', 'Room 4', 'Room 5', 'Room 6']
for i, (lower, upper, error) in enumerate(zip(lower_bound, upper_bound, relative_errors)):
    print(f"{room_labels[i]}: Mean:{means[i]}, 95% Confidence Interval: [{lower:.3f}, {upper:.3f}] (Relative Error: {error:.4f})")
```
