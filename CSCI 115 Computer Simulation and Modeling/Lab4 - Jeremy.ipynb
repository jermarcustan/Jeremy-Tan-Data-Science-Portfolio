### Government Queues
### Problems 1,2

```python
class CorruptQueue():
    
    def __init__(self, filename = None):
        self.filename = filename
        self.reglist = []
        self.vipstack = []
        self.supervisor_status = "leave"
        self.Lines = [] 
        
    def simulate(self):
        with open(self.filename) as f:
            lines = f.readlines()
            for line in lines:
                line = line.strip('\n')
                line = line.split(',')
                self.Lines.append(line)
                
        for i in self.Lines:
            if len(i) == 3:
                if self.supervisor_status == 'arrive':
                    self.reglist.append(i[1])
                    if i[2] == "regular":
                        print(f'Regular client {i[1]} lines up at RegularQueue') 
                    if i[2] == "VIP":
                        print(f'VIP client {i[1]} lines up at RegularQueue') 
                elif self.supervisor_status == 'leave':
                    if i[2] == 'regular':
                        self.reglist.append(i[1])
                        print(f'Regular client {i[1]} lines up at RegularQueue') 
                    elif i[2] == 'VIP':
                        self.vipstack.append(i[1])
                        print(f'VIP client {i[1]} lines up at VIPStack')
                        
            elif len(i) == 2:
                if i[0] == "arrive":
                    self.reglist = self.reglist + self.vipstack
                    self.vipstack.clear()
                    print('Supervisor present')
                    self.supervisor_status = i[0]
                elif i[0] == "leave":
                    print('Supervisor not here')
                    self.supervisor_status = i[0]

            elif len(i) == 1:
                if self.supervisor_status == 'arrive':
                    name = self.reglist.pop(0)
                    print(f'Now serving {name} from RegularQueue')
                if self.supervisor_status == 'leave':
                    if self.vipstack:
                        name = self.vipstack.pop()
                        print(f'Now serving {name} from VIPStack')
                    else:
                        name = self.reglist.pop(0)
                        print(f'Now serving {name} from RegularQueue')
```
