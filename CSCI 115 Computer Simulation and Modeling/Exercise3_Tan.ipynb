```python
population = 100
growth_rate = 0.1
time_step = 0.005
```

```python
import numpy as np
```

## Question 1
Simulate and evaluate: <br>
A. population(0.050h), the population at the next time interval after the end of Table of Estimated Populations above, to six decimal places. <br>
B. population(20.000h) to six decimal places. <br>

### Problem A

```python
population_array_1 = np.array([100])
time_array_1 = np.array([0])
for n in range(0, 10):
    current_population_1 = population_array_1[n] + population_array_1[n]*growth_rate*time_step
    population_array_1 = np.append(population_array_1, [current_population_1])
    current_time_1 = time_array_1[n] + 0.005
    time_array_1 = np.append(time_array_1, [current_time_1])
print(f"The population of bacteria after 0.050 hours is {round(population_array_1[-1], 6)}.")
```

### Problem B

```python
population_array_2 = np.array([100])
time_array_2 = np.array([0])
for n in range(0, 4000):
    current_population_2 = population_array_2[n] + population_array_2[n]*growth_rate*time_step
    population_array_2 = np.append(population_array_2, [current_population_2])
    current_time_2 = time_array_2[n] + 0.005
    time_array_2 = np.append(time_array_2, [current_time_2])
print(f"The population of bacteria after 20 hours is {round(population_array_2[-1], 6)}.")
```

## Question 2
Present the updated estimated population table vs time graph from t = 0.000h to 20.000h, using âˆ†t=0.005h.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(time_array_2, population_array_2)
ax1.set_xlabel("time (hours)")
ax1.set_ylabel("population")
ax1.set_title("Bacteria Population over 20 hours")
```

## Question 3
Discuss the trend of population growth from t=0.000h to 20.000h. Provide your analysis and insight with regard to the resulting trend.

<b><u>Analysis</u></b><br>
I noticed that the number of bacteria is increasing exponentially as the population growth of the bacteria continually increases  over time. The population graph does not increase linearly, but instead, the slope or the growth increases over time. The  results of this model are consistent with nature since as bacteria are dividing into new cells, as time passes, these new cells will result in even more cells undergoing binary fission. This model does not take into account the carrying capacity of the environment or other limiting factors; thus, the number of bacteria will increase without bound over time. 

To show that the population growth is truly increasing, I graphed the population growth over time with an increment of 0.005 hours (20 hours).

```python
growth_array = np.array([10])
population_array_2 = np.array([100])
for n in range(0, 4000):
    current_population_2 = population_array_2[n] + population_array_2[n]*growth_rate*time_step
    population_array_2 = np.append(population_array_2, [current_population_2])
    growth = population_array_2[n]*growth_rate
    growth_array = np.append(growth_array, [growth])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(time_array_2, growth_array)
ax1.set_xlabel("time (hours)")
ax1.set_ylabel("growth")
ax1.set_title("Bacteria Growth per hour ")
print(growth_array[-1])
print("The bacteria growth went from 0 to 10 at t=0 since we set the initial growth of the bacteria to be at 10 per hour")
```

The bacteria population is growing exponentially as the growth of bacteria  per hour increased from 10 to 73.81 after 20 hours.
