## OneShot Neural Network

```python
# Heart Disease Classsification Dataset

import pandas as pd

df = pd.read_csv("processed.cleveland.data", header = None)
df.replace('?', 0, inplace=True)
column_names = [f"x{i}" for i in range(1, 14)]
column_names.append("y")
df.columns = column_names
```

```python
pd.set_option('display.max_rows', None)  # or use a specific number
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
# get 1 observation from each class 

new_df = df.iloc[[0,1,2,6,24], :]
new_df
```

```python
import os

import keras.backend as K
from keras.models import Model, Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Input, Subtract, Lambda
from keras.optimizers import Adam, SGD
from keras.regularizers import l2
import keras.backend as K

import tensorflow as tf

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
from itertools import combinations_with_replacement
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models, Input
from sklearn.model_selection import train_test_split


X = new_df[['x1', 'x2', 'x3', 'x4', 'x5', 'x6', 'x7', 'x8', 'x9', 'x10', 'x11', 'x12', 'x13']].values
y = new_df['y'].values

# Generate the pairs for the neural network input 
# Since we are allowed to reuse the same observation, we have 15 total pairs
# The label for the pair is 1 if they are from the same class, 0 otherwise.

indices = list(combinations_with_replacement(range(len(X)), 2))
pairs = [(X[i], X[j]) for i, j in indices]
labels = []

# Compare the classes using y
for i, j in indices:
    if y[i] == y[j]:
        labels.append(1)  # Same class
    else:
        labels.append(0)  # Different class
        
# Split data into training and test sets
pairs_train, pairs_test, labels_train, labels_test = train_test_split(pairs, labels, test_size=0.2, random_state=42)

```

```python
# We have to convert the train and test lists to numpy arrays
# Convert pairs to arrays and ensure data types are floats
pairs_train_1 = np.array([[float(value) for value in pair[0]] for pair in pairs_train])
pairs_train_2 = np.array([[float(value) for value in pair[1]] for pair in pairs_train])
pairs_test_1 = np.array([[float(value) for value in pair[0]] for pair in pairs_test])
pairs_test_2 = np.array([[float(value) for value in pair[1]] for pair in pairs_test])

# Convert labels to arrays and ensure data types are integers
labels_train = np.array(labels_train, dtype=np.int)
labels_test = np.array(labels_test, dtype=np.int)
```

```python

# Siamese network architecture
def create_base_network(input_shape):
    model = models.Sequential()
    model.add(layers.Dense(128, activation='relu', input_shape=input_shape))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(128, activation='relu'))
    return model

input_shape = (13,)

# Build the subnet
base_network = create_base_network(input_shape)

# Inputs for the two subnetworks
input_a = Input(shape=input_shape)
input_b = Input(shape=input_shape)

# Get the embeddings for both inputs
processed_a = base_network(input_a)
processed_b = base_network(input_b)
```

```python
# Compute the L1 distance between the two embeddings
distance = layers.Lambda(lambda x: tf.abs(x[0] - x[1]))([processed_a, processed_b])

# Output layer: binary classification (same or different class)
'''
 Since the sigmoid activation function was used, the output of the predict function 
 will be the probability that the two inputs are of the same class.
'''
output = layers.Dense(1, activation='sigmoid')(distance)

# Define the Siamese model
siamese_net = models.Model([input_a, input_b], output)

# Compile the model
# The paper used cross-entropy loss.
siamese_net.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
siamese_net.fit([pairs_train_1, pairs_train_2], labels_train, 
                validation_data=([pairs_test_1, pairs_test_2], labels_test), 
                epochs=10, batch_size=64)


# Evaluate the model
results = siamese_net.evaluate([pairs_test_1, pairs_test_2], labels_test)
print(f'Test Loss: {results[0]}, Test Accuracy: {results[1]}')
```

```python
index = 4 # Choose a data point in the dataframe which is not part of the test set.
sample_a1 = np.array(df.iloc[index, :13], dtype = np.float32)
print(f'The class label for the test instance is {df.iloc[index, -1]}.')

# Create pairs with the training examples for each class

sample_b1 = pairs_train_1[3] # class y = 0
sample_b2 = pairs_train_2[3] # class y = 1
sample_b3 = pairs_train_2[1] # class y = 2
sample_b4 = pairs_train_2[7] # class y = 3
sample_b5 = pairs_train_2[0] # class y = 4

pairs_1 = np.array([[sample_a1, sample_b1]]) # class 0
pairs_2 = np.array([[sample_a1, sample_b2]]) # class 1
pairs_3 = np.array([[sample_a1, sample_b3]]) # class 2
pairs_4 = np.array([[sample_a1, sample_b4]]) # class 3
pairs_5 = np.array([[sample_a1, sample_b5]]) # class 4

# Use the model to predict the probability that the new data point is of a particular class
# We input the data point pair into the predict function of the NN.
probability_1 = siamese_net.predict([pairs_1[:, 0], pairs_1[:, 1]])
probability_2 = siamese_net.predict([pairs_2[:, 0], pairs_2[:, 1]])
probability_3 = siamese_net.predict([pairs_3[:, 0], pairs_3[:, 1]])
probability_4 = siamese_net.predict([pairs_4[:, 0], pairs_4[:, 1]])
probability_5 = siamese_net.predict([pairs_5[:, 0], pairs_5[:, 1]])


predictions = [probability_1, probability_2, probability_3, probability_4, probability_5]

'''
The class label pair with the highest probability is the network's prediction 
of which class does the test data point most likely belong to.
'''

print(f'The predicted class label for the test instance is {np.argmax(predictions)}.')
```

<h3><b>Guide Questions:</b></h3>

1. What were the activation functions you used in the hidden layer and output layer? Why were these the appropriate activation functions for this neural network?

<p>The RELU activation function was used for the hidden layer while the sigmoid function was used for the output layer. The RELU activation function introduces sparsity into the word embeddings, leading to a more compact representation of the datapoints. Moreover, the RELU activation function allows the neural network to efficiently learn through backpropagation since the gradients don't shrink when RELU is used. The sigmoid function in the output layer maps the distance between the embedding vectors of the two data points onto the scale of 0 to 1, which represents the similarity or probability that the two images come from the same class.</p>

2. How does a siamese neural network perform classification? (in your own words)<br>

<p>The siamese NN performs oneshot classification by training on pairs of data points. The NN accepts inputs of pairs of observations with the same and different classes. The goal of the siamese NN is to predict the degree of similarity between any two data points. </p>

<p>In essence, the siamese NN is able to classify new points despite having less training examples because it merely compares the new instance with all of the examples it has trained on. The network then selects the class which is the most similar to the test instance as its prediction.</p>
