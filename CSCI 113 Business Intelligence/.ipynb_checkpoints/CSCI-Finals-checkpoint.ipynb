Looking at the values of the columns, we can remove the columns: EmployeeCount, StandardHours, and Over18 since these features are not relevant in determining whether the employee stays or leaves the company. 

```python
clean = clean.rename(clean["EmployeeNumber"], axis = 0) # Changes the index to the employee number
attrition_col = clean["Attrition"]
attrition_col = attrition_col.eq('Yes').mul(1)
clean.drop(["EmployeeNumber","EmployeeCount", "StandardHours", "Over18", "Attrition"], inplace = True, axis = 1)
```

## Data Preprocessing

```python
#categorical
cat_df = pd.DataFrame(clean, columns=clean.columns[(clean.dtypes == 'object') | (clean.dtypes == 'category')])
cat_df = pd.get_dummies(cat_df, dtype = int)
#continuous
cont_df = pd.DataFrame(clean, columns = clean.columns[(clean.dtypes == 'int64')])
normalized_df = (cont_df-cont_df.min())/(cont_df.max()-cont_df.min())
processed_clean = pd.concat([normalized_df,cat_df],axis=1)
```

```python
# Inserting the Attrition column

processed_clean.insert(0, "Attrition", attrition_col)
cat_df.insert(0, "Attrition", attrition_col)
normalized_df.insert(0, "Attrition", attrition_col)
```

```python
skipped = []
def binwidth(data):
  '''Given a certain pandas Series or array of numerical data, identify the IQR, and return the bin_width given by the Freedman - Diaconis Rule'''
  q1, q3 = np.percentile(data, [25,75]) ## identify 25th and 75th quartiles
  iqr = q3 - q1 ## calculating for iqr
  n = len(data) ## identify number of rows
  bin_width = (2*iqr)/n**(1/3) ## The Freedman - Diaconis Rule is rounded by default math rounding rules
  return bin_width ## return bin_width

for column in processed_clean.columns[1:]: 
  print(column)
  retention2 = retention.loc[:, column] 
  attrition2 = attrition.loc[:, column] 

  if binwidth(processed_clean[column]) != 0: # remove 
    bins = np.arange(min(processed_clean[column]), max(processed_clean[column]) + binwidth(processed_clean[column]), binwidth(processed_clean[column]))

    plt.hist(retention2, alpha = 0.5, bins=bins, label = f"{column}_retention") ## Plotting a histogram for dataframe m
    plt.hist(attrition2, alpha = 0.5, bins=bins, label = f"{column}_attrition") ## Plotting a histogram for dataframe b
    plt.legend(loc = "upper right") ## Adding necessary Legend
    plt.show() ## Show Plot
  else:
    skipped.append(column)
print(skipped)
```
