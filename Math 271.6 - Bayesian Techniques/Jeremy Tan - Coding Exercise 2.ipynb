```python
import numpy as np
from scipy.stats import multivariate_normal, norm

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

import pandas as pd
```

```python
np.random.seed(42)
```

```python
df = pd.read_csv("telecom.csv")
```

```python
df
```

```python
X = df[['Average_SMS_per_day', 'Average_Call_Duration_per_day']].values  
X
```

## Gaussian Mixture (2 Clusters)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X)

means = gmm.means_
covariances = gmm.covariances_
weights = gmm.weights_

labels = gmm.predict(X)

print("Means:\n", means)
print("\nCovariances:\n", covariances)
print("\nMixture Weights:\n", weights)

# Visualization
plt.figure(figsize=(8, 6))

for cluster in range(2):
    cluster_points = X[labels == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                s=30, label=f'Cluster {cluster + 1}', cmap='viridis')

plt.scatter(means[:, 0], means[:, 1], c='red', marker='x', s=100, label='Cluster Centers')

plt.title("GMM Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

## Evaluating Clustering

```python
log_likelihood = gmm.score(X)
print("Mean Log-Likelihood:", log_likelihood)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X, labels)
print("Silhouette Score:", sil_score)
```

## Gaussian Mixture (3 Clusters)

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# Get the parameters
means = gmm.means_
covariances = gmm.covariances_
weights = gmm.weights_

# Predict the cluster for each data point
labels = gmm.predict(X)

# Print the distribution parameters
print("Means:\n", means)
print("\nCovariances:\n", covariances)
print("\nMixture Weights:\n", weights)

# Visualization
plt.figure(figsize=(8, 6))

# Plot each cluster individually with a label
for cluster in range(3):
    cluster_points = X[labels == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                s=30, label=f'Cluster {cluster + 1}', cmap='viridis')

# Plot the Gaussian centers
plt.scatter(means[:, 0], means[:, 1], c='red', marker='x', s=100, label='Cluster Centers')

plt.title("GMM Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```

## Compare Clustering (K = 2 to K = 10)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

log_likelihoods = []
silhouette_scores = []
ks = range(2, 11)

for k in ks:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(X)
    
    labels = gmm.predict(X)
    
    log_likelihood = gmm.score(X)
    log_likelihoods.append(log_likelihood)
    
    if len(np.unique(labels)) > 1:
        sil_score = silhouette_score(X, labels)
    else:
        sil_score = -1  
    silhouette_scores.append(sil_score)

plt.figure(figsize=(12, 5))

# Log-Likelihood plot
plt.subplot(1, 2, 1)
plt.plot(ks, log_likelihoods, marker='o')
plt.title("Mean Log-Likelihood vs Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Mean Log-Likelihood")

# Silhouette Score plot
plt.subplot(1, 2, 2)
plt.plot(ks, silhouette_scores, marker='o', color='green')
plt.title("Silhouette Score vs Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.tight_layout()
plt.show()

clustering_metrics = pd.DataFrame({
    'num_clusters': np.arange(2,11),
    'mean_log_likelihood': log_likelihoods,
    'silhouette_score': silhouette_scores
})

print(clustering_metrics.to_string())

best_k = ks[np.argmax(silhouette_scores)]
print("Best number of clusters based on silhouette score:", best_k)
```

<b>Optimal k (number of clusters)</b>

Based on both mean log likelihood and silhouette score, the optimal number of clusters for the dataset was three. The highest silhouette score, 0.766678, was obtained when k = 3, and this was significantly greater than the second highest silhouette score, 0.709455, when k = 2. When looking at k = 4, we can see that the mean log likelihood slightly increased from -5.851205 to -5.716092, but at the cost of significantly decreasing the silhouette score down to 0.587200. Thus, the optimal number of clusters for this GMM was three. 

<b>Questions</b>:

1. How do log-likelihood and silhouette score complement each other in
evaluating clustering performance
2. Does the highest silhouette score from k = 2 to 10 always indicate the
best number of clusters? Why or why not
?

<b>Answers</b>:

1. 
<p>Both metrics complement each other since they measure different aspects of the data clusters. The log-likelihood metric is a probabilistic measure of how well the data in each cluster fit a Gaussian distribution. As the number of clusters (K) increases, the log likelihood increases. However, this can lead to overfitting where near data points are assigned to different clusters.

Thus, it is important to balance the log likelihood metric with another metric, the silhouette score, which measures how well separated the clusters are. If the number of clusters (K) is too high, the silhouette score would be low since the data points in a cluster would be close to the datapoints of the closest neighbor cluster. Thus, using both metrics would ensure a proper balance between the fit of the data points and how well-separated each cluster is. </p>

2. 
<p>No, the highest silhouette score does not necessarily mean the number of clusters is optimal since we would also need to check the fit (log-likelihood) for that number of clusters. In the case of Gaussian Mixture Models (GMMs), the clusters are assumed to be spherical in shape. This could mean that the distance metric which is used in calculating the silhouette score may not take into account the density of the data points in complex shapes. Moreover, a probabilistic measure like log-likelihood or BIC is important in GMMs since the data is probabilistically assigned to each cluster. </p>
