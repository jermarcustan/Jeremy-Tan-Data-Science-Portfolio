```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

## Poisson Likelihood and Gamma Prior



### Part 1: Understanding the Prior

```python
# Plot the pdf of the gamma distribution
alpha_0 = 8
beta_0 = 1
x = np.linspace(0,30,1000)
y = gamma.pdf(x, a=alpha_0, scale=1/beta_0)

prior_mean = alpha_0 / beta_0
prior_variance = alpha_0 / beta_0**2

print(f'The mean of the prior gamma distribution is {prior_mean} while the variance is {prior_variance}.')

```

```python
plt.plot(x,y)
plt.show()
```

### Part 2: Bayesian Updating with Observed Data

```python
records = [10,7,12]

alpha_prior = 8
beta_prior = 1

x = np.linspace(0,30,1000)
y = gamma.pdf(x, a=alpha_0, scale=1/beta_0)
plt.plot(x,y, label = f"Prior Distribution")

alpha_post = alpha_prior
beta_post = beta_prior

for i in range(3):

    alpha_post += records[i]
    beta_post += 1
    print(f'After {i+1} day(s), alpha = {alpha_post} and beta = {beta_post}.')
    y = gamma.pdf(x, a = alpha_post, scale = 1/beta_post)
    plt.plot(x,y, label = f"Posterior after {i+1} record")
    
plt.legend()
plt.show()

print(f'The posterior mean is {alpha_post / beta_post}.')
    
```
