```python
import pandas as pd
import numpy as np
import os
```

## Dataset

```python
BASE_DIR_PATH = '../Datasets'
DATASET_FILE = os.path.join(BASE_DIR_PATH, '2016_2017_sto_nino_and_montalban.csv')
```

```python
df = pd.read_csv(DATASET_FILE)
df
```

```python
df = df[['month', 'day', 'hour', 'Station', 'Rainfall_Aries', 'Rainfall_Boso', 'Rainfall_Campana', 'Rainfall_Nangka', 'Rainfall_Oro', 'Waterlevel']]
df
```

```python
LAG = 12
lagged_df = df.copy()

for i in range(LAG):
    j = i+1
    col_name = f'Waterlevel_t-{j}'
    lagged_data = df['Waterlevel'].shift(j*2)
    lagged_data = lagged_data.rename(col_name)
    
    lagged_df = pd.concat([lagged_df, lagged_data], axis=1)

lagged_df
```

```python
lagged_df.to_csv(f'../Datasets/2016_2017_sto_nino_and_montalban_lag_{int(LAG)}.csv')
```
