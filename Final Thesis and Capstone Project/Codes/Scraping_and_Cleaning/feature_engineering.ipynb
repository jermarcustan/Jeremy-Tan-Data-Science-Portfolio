```python
import numpy as np

Y = np.array(sto_nino_rating_curve_train['Waterlevel_Sto_Nino'])
X = np.array(sto_nino_rating_curve_train['Discharge_Sto_Nino'])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y)
plt.show()
```

### Using Power Law Curve-fitting

```python
from scipy.optimize import curve_fit

def f(x, a, b, c):
    return a * (x - b) ** c
```

```python
popt, pcov = curve_fit(f, Y, X)

a, b, c = popt
a, b, c
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, label='Data')
plt.scatter(a * (Y-b) **c, Y, color='red', label='Rating Curve')
plt.xlabel('Water Discharge')
plt.ylabel('Water Level')
plt.title('Rating Curve for Sto. Nino')
plt.legend()
plt.show()
```

```python
montalban_wl = final_df[['Date', 'Time', 'Waterlevel_Montalban']]
montalban_wl
```

```python
montalban_daily_ave = montalban_wl.groupby('Date').mean()
montalban_daily_ave = montalban_daily_ave.reset_index()
montalban_daily_ave
```

```python
montalban_rating_curve = montalban_daily_ave.merge(discharge_df[['Discharge_San_Jose', 'Date']], on='Date')
montalban_rating_curve
```

```python
montalban_rating_curve.info()
```

```python
Y2 = np.array(montalban_rating_curve['Waterlevel_Montalban'])
X2 = np.array(montalban_rating_curve['Discharge_San_Jose'])
```

```python
popt2, _ = curve_fit(f, Y2, X2, maxfev=100000)

a2, b2, c2 = popt2
a2, b2, c2
```

```python
plt.scatter(X2, Y2, label='Data')
plt.scatter(a2 * (Y2-b2) **c2, Y2, color='red', label='Rating Curve')
plt.xlabel('Water Discharge')
plt.ylabel('Water Level')
plt.title('Rating Curve for Montalban')
plt.legend()
plt.show()
```

## Solving for Discharge

```python
final_df['Discharge_Sto_Nino'] = a*(np.array(final_df['Waterlevel_Sto_Nino'])-b)**c
final_df
```

```python
final_df.info()
```

```python
final_df['Discharge_San_Jose'] = a2*(np.array(final_df['Waterlevel_Montalban'])-b2)**c2
final_df
```
