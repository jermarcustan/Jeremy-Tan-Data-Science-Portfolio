```python
import pandas as pd
import os
```

```python
BASE_DIR_PATH = '/Users/neilb/Documents/dsci_thesis/Datasets'
DATASET_FILE = os.path.join(BASE_DIR_PATH, 'compiled_data_2016_2017.csv')
```

```python
df = pd.read_csv(DATASET_FILE)
df
```

```python
df = df.iloc[:,2:]
df
```

```python
df_Sto_Nino = df[['datetime', 'x', 't', 'Rainfall_Aries', 'Rainfall_Boso', 'Rainfall_Campana', 'Rainfall_Nangka', 'Rainfall_Oro', 'Waterlevel_Sto_Nino', 'Discharge_Sto_Nino', 'Cross_Section_Sto_Nino', 'Velocity_Sto_Nino']]
df_Sto_Nino.head()
```

```python
df_Sto_Nino = df_Sto_Nino.rename(columns={
    'Waterlevel_Sto_Nino': 'Waterlevel', 
    'Discharge_Sto_Nino': 'Discharge', 
    'Cross_Section_Sto_Nino': 'Cross_Section', 
    'Velocity_Sto_Nino': 'Velocity'
})
df_Sto_Nino
```

```python
df_Sto_Nino['Station'] = ['Sto_Nino' for x in range(len(df_Sto_Nino))]
df_Sto_Nino['fric_coeff'] = [0.033 for x in range(len(df_Sto_Nino))]
df_Sto_Nino['slope'] = [1/1500 for x in range(len(df_Sto_Nino))]
df_Sto_Nino
```

```python
df_Montalban = df[['datetime', 'x', 't', 'Rainfall_Aries', 'Rainfall_Boso', 'Rainfall_Campana', 'Rainfall_Nangka', 'Rainfall_Oro', 'Waterlevel_Montalban', 'Discharge_San_Jose', 'Cross_Section_Montalban', 'Velocity_Montalban']]
df_Montalban.head()
```

```python
df_Montalban = df_Montalban.rename(columns={
    'Waterlevel_Montalban': 'Waterlevel', 
    'Discharge_San_Jose': 'Discharge', 
    'Cross_Section_Montalban': 'Cross_Section', 
    'Velocity_Montalban': 'Velocity'
})
df_Montalban
```

```python
df_Montalban['Station'] = ['Montalban' for x in range(len(df_Sto_Nino))]
df_Montalban['fric_coeff'] = [0.035 for x in range(len(df_Sto_Nino))]
df_Montalban['slope'] = [0.002 for x in range(len(df_Sto_Nino))]
df_Montalban['x'] = [0 for x in range(len(df_Montalban))]
df_Montalban
```

```python
final_df = pd.concat([df_Sto_Nino, df_Montalban], axis=0)
final_df
```

```python
final_df = final_df[['datetime', 'Station', 'x', 't', 'Rainfall_Aries', 'Rainfall_Boso', 'Rainfall_Campana', 'Rainfall_Nangka', 'Rainfall_Oro', 'Waterlevel', 'Discharge', 'Cross_Section', 'Velocity', 'fric_coeff', 'slope']]
final_df.sort_values(['datetime', 'Station'], inplace=True)
final_df.reset_index(drop=True, inplace=True)
final_df
```

```python
final_df.info()
```

```python
final_df['datetime'] = pd.to_datetime(final_df['datetime'])
final_df.info()
```

```python
final_df['hour'] = [x.hour for x in final_df['datetime']]
final_df
```

```python
final_df = final_df[['datetime', 'month', 'day', 'hour', 't', 'Station', 'x', 'Rainfall_Aries', 'Rainfall_Boso', 'Rainfall_Campana', 'Rainfall_Nangka', 'Rainfall_Oro', 'Waterlevel', 'Discharge', 'Cross_Section', 'Velocity', 'fric_coeff', 'slope']]
final_df
```
