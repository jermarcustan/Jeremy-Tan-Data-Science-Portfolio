```python
# import library stuff for plotting
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import new dataset for plotting (table_i702t60.csv)
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['table_i702t60.csv']))
```

```python
df.head()
```

```python
df.info()
```

```python
# get values for the 3 rows to plot
month_values = df['month'].values
interest_paid = df['interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```python
month_values
```

```python
# showing interest paid over time
plt.plot(month_values, interest_paid)
```

```python
# showing principal paid over time
plt.plot(month_values, principal_paid)
```

```python
# show available plot styles
plt.style.available
```

```python
# lets combine and style it
plt.style.use('seaborn')
plt.plot(month_values, interest_paid)
plt.plot(month_values, principal_paid)
```

```python
# with custom markers and colors in matlab syntax
plt.style.use('seaborn')
plt.plot(month_values, interest_paid, marker='.', markersize=10, color='k')
plt.plot(month_values, principal_paid, marker='*', markersize=10, color='b')
```

```python
# with custom markers and colors in object oriented syntax
plt.style.use('seaborn')
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_values, interest_paid, marker='.', markersize=10, color='k', label='interest')
axes.plot(month_values, principal_paid, marker='*', markersize=10, color='b', label='principal')
plt.legend()
```

```python
plt.style.use('seaborn')
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(month_values, interest_paid, marker='.', markersize=10, color='k', label='interest')
axes[0].legend()
axes[1].plot(month_values, principal_paid, marker='*', markersize=10, color='b', label='principal')
axes[1].legend()
```

```python
# with custom markers and colors in matlab syntax
plt.style.use('seaborn')
plt.plot(month_values, interest_paid, marker='.', markersize=10, color='k', label="Interest")
plt.plot(month_values, principal_paid, marker='*', markersize=10, color='b', label="Principal")
plt.grid(axis='x')
plt.xlim(left=1,right=70)
plt.ylim(top=800,bottom=0)
plt.xlabel('Month')
plt.ylabel("Dollars")
plt.title("Interest and Principal Payments per Month", fontsize=20)
plt.legend()
```

```python
# import library stuff for plotting
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import new dataset for plotting
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['wisconsinBreastCancer.csv']))
```

```python
df.head()
```

```python
df.info()
```

```python
# See possible diagnosis results
df.diagnosis.value_counts(dropna=False)
```

```python
# generate boxplot for malignant and benign diagnosis but only the area_mean column
benign = df.loc[df['diagnosis']=='B', "area_mean"].values
malignant = df.loc[df['diagnosis']=='M', "area_mean"].values

plt.boxplot([malignant,benign], labels=['M','B'])
```

```python
# same plot using less code
df.boxplot(column="area_mean", by="diagnosis")
plt.title('')
plt.suptitle('')
```

```python
# same plot using seaborn
sns.boxplot(x='diagnosis', y='area_mean', data=df)
```

Visualization (Subplot) using Digits Data



```python
# import library stuff for plotting
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import new dataset for plotting
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(io.BytesIO(uploaded['digitsDataset.csv']))
```

```python
df.head()
```

```python
df.info()
```

```python
pixels_colnames = df.columns[:-1]
```

```python
pixels_colnames
```

```python
images_values = df.loc[0, pixels_colnames].values
```

```python
images_values
```

```python
images_values.shape
```

```python
images_values.reshape(8,8)
```

```python
# lets visual the dataset for one number
plt.imshow(images_values.reshape(8,8),cmap='gray')
```

```python
plt.imshow((df.loc[1, pixels_colnames].values).reshape(8,8),cmap='gray')
```

```python
plt.imshow((df.loc[2, pixels_colnames].values).reshape(8,8),cmap='gray')
```

```python
# lets built our subplot
plt.figure(figsize=(10,4))
plt.style.use('seaborn')

# this is generates the images over a loop
for index in range(0, 10):
  plt.subplot(2, 5, 1 + index )
  image_values = df.loc[index, pixels_colnames].values
  image_label = df.loc[index, 'label']
  plt.imshow(image_values.reshape(8,8), cmap ='gray')
  plt.title('Label: ' + str(image_label))
```
