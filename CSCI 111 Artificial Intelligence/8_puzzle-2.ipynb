**1st Search Algorithm: Breadth First Search (BFS)**


*   An uninformed search strategy where all children nodes are examined at each level from the initial state of the board.
*   For the 8-puzzle problem, we have a branching factor equal to the number of tiles adjacent to the empty tile.
*   If the tile is in the middle, we have 4 children nodes. If its in the edge, 3 children nodes. If its in the corner, we have 2 children nodes.
*   The BFS search algorithm will take note of past actions to avoid repetitive states.





```python
import numpy as np
```

```python
# Function inputs: our initial state and the goal state 
initial_list = np.array([[3,1,2], [4,5,8], [6,7,"_"]])
goal_state = np.array([["_", 1,2], [3,4,5], [6,7,8]])
```

```python
# Set-up functions to print the board, getting index, switching the board, etc. 
def print_board(initial_list):
  for row in initial_list:
    print(f"{row[0]} | {row[1]} | {row[2]}")
  print("\n")
print_board(initial_list)

def getting_index_of_empty(initial_list):
  location = np.where(initial_list == "_" )
  coordinates = []
  for coordinate in location:
    coordinate = coordinate.tolist()
    for x in coordinate:
      coordinates.append(x)
  return coordinates # row column (index starting with 0)

def getting_adjacent_tiles(initial_list):
  neighboring_array = []
  coordinates = getting_index_of_empty(initial_list)
  up = [coordinates[0]-1, coordinates[1]]
  down = [coordinates[0]+1, coordinates[1]]
  left = [coordinates[0], coordinates[1]-1]
  right = [coordinates[0], coordinates[1]+1]
  neighboring_array.extend([up, down, left, right])
  filters = [3, -1]
  final_l = [i for i in neighboring_array if not any(b in filters for b in i)]
  neighbors = []
  for num in final_l:
    x = num[0]
    y = num[1]
    a = initial_list[x,y]
    neighbors.append(a)
  return final_l, neighbors

def switch_board(initial_list):
  boards = []
  boards.append(initial_list)
  empty_tile_coordinates = getting_index_of_empty(initial_list)
  neighboring_coordinates, neighbors = getting_adjacent_tiles(initial_list)
  for count, coordinate in enumerate(neighboring_coordinates):
    swapped_list = np.copy(initial_list)
    x = coordinate[0]
    y = coordinate[1]
    swapped_list[x,y] = "_"
    empty_tile_coordinate_x = empty_tile_coordinates[0]
    empty_tile_coordinate_y = empty_tile_coordinates[1]
    swapped_list[empty_tile_coordinate_x, empty_tile_coordinate_y] = neighbors[count]
    boards.append(swapped_list)
  return boards[1:]

def inList(test, goal):
    for element in test:
        if np.array_equal(element, goal):
            return True
    return False



```

```python
# Breadth First Search Algorithm
def BFS(initial_list, goal_state, flag): 
  flag = flag # When getting the path, we set flag to 1 to not print the search tree multiple times 
  if flag == 0:
    print("Our Initial State is:")
    print(initial_list)
  queue = [initial_list] # contains all of the nodes we have explored
  checker = True 
  for i in range(20000): # to prevent infinite loops
    if checker == False: 
      break
    if flag == 0:
      print(f'The children of the {i}th node expanded')
    if inList(queue, goal_state):
      print(queue[i])
      break
    temp_queue = switch_board(queue[i])
    for node in temp_queue:
      if (node==goal_state).all():
        if flag == 0:
          print("We have reached the goal node!")
        x = i #for backtracking, x is the i at which we found the goal node!
        checker = False
      if inList(queue, node) == False: # We don't revisit children nodes we have already explored
        if flag == 0:
          print(node)
        queue.append(node)
  return x, queue

BFS(initial_list, goal_state, 0)
```

```python
# Returns the correct path from initial state to goal state.
def path(goal, initial):
  flag = 1
  parent_node = goal
  steps = [parent_node]
  while True:
    if (parent_node == initial).all():
      break
    x, queue = BFS(initial_list, parent_node, flag)
    parent_node = queue[x]
    steps.append(parent_node) 
  inverted_steps = steps[::-1]
  for count, array in enumerate(inverted_steps):
    if count == 0:
      print("Initial State")
    else:
      print(f'Move number: {count}')
    print_board(array)

path(goal_state,initial_list)
```

**2nd Search Algorithm: A* Search**


*   Informed Search Algorithm based on a cost and distance heuristic.
*   Our evaluation function is f(n) = g(n) + h(n). <br>
<p>g(n) is the cost to reach the node. We measure it by the depth of the node on the search tree. h(n) is the distance heurstic from current noad to the goal node. We are going to measure both Manhattan and Hamming distances. </p>



*   Manhattan distance is the sum of the right angle distances between the tile and its correct position while Hamming distance is the number of misplaced tiles.


*   Our algorithm picks the node with the <b>least</b> f score and expands it.














```python
# Set up Functions to calculate the distances, get the coordinates, find the parent node, etc. 
def getting_index_of_key(current_state, key):
  location = np.where(current_state == key)
  coordinates = []
  for coordinate in location:
    coordinate = coordinate.tolist()
    for x in coordinate:
      coordinates.append(x)
  return coordinates # row column (index starting with 0)

def get_coordinates_of_all(current_state):
  number_dict = {"_": getting_index_of_key(current_state, "_")}
  for i in range (1,9):
    i_str = str(i)
    current_key =  getting_index_of_key(current_state, i_str)
    number_dict[i] = current_key
  return number_dict

from math import sqrt

#create function to calculate Manhattan distance 
def manhattan(a, b):
    sum = 0
    for val1, val2 in zip(a,b):
        for a, b in zip(val1,val2):
          sum+=abs(a-b)
    return sum
 
def hamming(a,b): # number of misplaced tiles
  sum = 0
  for val1, val2 in zip(a,b):
    if val1 != val2:
      sum += 1
  return sum

def parent_finder(paths, state):
  for i, node in enumerate(paths):
    if (node[-1] == state).all():
      return paths[i]

```

```python
initial_list = np.array([[1,8,2], ["_",4,3], [7,6,5]])
goal_state = np.array([["_", 1,2], [3,4,5], [6,7,8]])
```

```python
# A* search algorithm with either Manhattan or Hamming distance heuristic
def a_star(initial_list, goal_state, flag, heuristic): 
  heuristic = heuristic
  # Our priority queue will contain the initial state and the f score of the board
  priority_queue = {}

  l = 0 # for each child node
  paths = [[initial_list]] # store the info on the paths here

  # We will also have a list containing all the states we have explored.
  states = [initial_list]
  if flag == 0:
    print("Initial State:")
    print(states[0])
  g_score = 0 # depth of initial node is 0
  if heuristic == "Manhattan":
    h_score = manhattan(get_coordinates_of_all(initial_list).values(), (get_coordinates_of_all(goal_state).values()))
  elif heuristic == "Hamming":
    h_score = hamming(get_coordinates_of_all(initial_list).values(), (get_coordinates_of_all(goal_state).values()))
  f_score = g_score + h_score
  # convert numpy array to a tuple because the key of a python dictionary must be unhashable
  initial_list = tuple(map(tuple, initial_list)) 
  priority_queue[initial_list] = f_score 
  cost = 0 # cost is equivalent to the depth of the search tree
  checker = True 
  for i in range(20000): # to prevent infinite loops
    if checker == False: 
      break
    if flag == 0:
      print(f'The children of the {i}th node expanded')
    lowest = min(priority_queue.items(), key=lambda x: x[1]) # get the key-value pair with the lowest f score
    state = lowest[0]
    priority_queue.pop(state) # pop/remove the state from the priority queue dictionary
    state = np.asarray(state) # convert tuple back to a numpy array
    parent_path = parent_finder(paths, state)
    min_f_score = lowest[1]
    if (state==goal_state).all():
      if flag == 0:
        print(state)
      break
    temp_queue = switch_board(state)
    for node in temp_queue:
      l+=1
      children_path = parent_path.copy()
      children_path.append(node)
      paths.append(children_path)
      if (node==goal_state).all():
          if flag == 0:
            print("We have reached the goal node!")
          checker = False
          correct_path = paths[l] # retrieve the correct path when we reach the goal node
      if inList(states, node) == False: # We only explore the children nodes we haven't explored before
        states.append(node)
        if heuristic == "Manhattan":
          h_score = manhattan(get_coordinates_of_all(node).values(), (get_coordinates_of_all(goal_state).values()))
          parent_g_score = min_f_score - manhattan(get_coordinates_of_all(state).values(), (get_coordinates_of_all(goal_state).values()))
        elif heuristic == "Hamming":
          h_score = hamming(get_coordinates_of_all(node).values(), (get_coordinates_of_all(goal_state).values()))
          parent_g_score = min_f_score - hamming(get_coordinates_of_all(state).values(), (get_coordinates_of_all(goal_state).values()))
        child_g_score = parent_g_score + 1 # We add one to the g score (depth) of the parent node to get the child g score
        new_f_score = child_g_score + h_score
        if flag == 0:
          print(f'g-score = {child_g_score} --- h_score = {h_score}')
          print(node)
        node = tuple(map(tuple, node))
        priority_queue[node] = new_f_score 
  return correct_path

a_star(initial_list, goal_state, 0, "Manhattan")



```

```python
# Generates and outputs the correct path for the A* search algorithm
initial_list = np.array([[1,8,2], ["_",4,3], [7,6,5]])
correct_path = a_star(initial_list, goal_state, 1, "Manhattan")
def a_star_path(correct_path):
  count = 0
  for array in correct_path:
    if count == 0:
      print("Initial State")
    else:
      print(f'Move number: {count}')
    print_board(array)
    count += 1
print("\n")
print("Correct Path ")
print("\n")
a_star_path(correct_path)
```

**Greedy Best-first search**



*   A simplified version of A* search which only takes into account a single heuristic h(n), which is a distance function. 

*   The distance function can either be Manhattan distance or Hamming Distance between the current state and the goal state. 







```python
hard_example = np.array([[1,8,2], ["_",4, 3], [7,6,5]])
goal_state = np.array([["_", 1, 2], [3,4,5], [6,7,8]])
```

```python
# Greedy Best First Search search algorithm with either Manhattan or Hamming distance heuristic
hard_example = np.array([[1,8,2], ["_",4, 3], [7,6,5]])
def best_first(initial_list, goal_state, flag, heuristic): 
  flag = flag # We only print the search tree once
  heuristic = heuristic
  
  # Our priority queue will contain the initial state and the h score of the board
  priority_queue = {}

  l = 0 # for each child node
  paths = [[initial_list]] # store the info on the paths here
  if flag == 0:
    print(initial_list)
  # We will also have a list containing all the states we have explored.
  states = [initial_list]
  if heuristic == "Manhattan":
    h_score = manhattan(get_coordinates_of_all(initial_list).values(), (get_coordinates_of_all(goal_state).values()))
  elif heuristic == "Hamming":
    h_score = hamming(get_coordinates_of_all(initial_list).values(), (get_coordinates_of_all(goal_state).values()))
  # convert numpy array to a tuple because the key of a python dictionary must be unhashable
  initial_list = tuple(map(tuple, initial_list)) 
  priority_queue[initial_list] = h_score 
  cost = 0 # cost is equivalent to the depth of the search tree
  checker = True 
  for i in range(20000): # to prevent infinite loops
    if checker == False: 
      break
    if flag == 0:
      print(f'The children of the {i}th node expanded')
    lowest = min(priority_queue.items(), key=lambda x: x[1]) # get the key-value pair with the lowest h score
    state = lowest[0]
    priority_queue.pop(state)
    state = np.asarray(state) # convert tuple back to a numpy array
    parent_path = parent_finder(paths, state)
    min_h_score = lowest[1]
    if (state==goal_state).all():
      if flag == 0:
        print(state)
      break
    temp_queue = switch_board(state)
    for node in temp_queue:
      l+=1
      children_path = parent_path.copy()
      children_path.append(node)
      paths.append(children_path)
      if (node==goal_state).all():
        if flag == 0:
          print("We have reached the goal node!")
        checker = False
        correct_path = paths[l]
      if inList(states, node) == False: # We only explore the children nodes we haven't explored before
        states.append(node)
        if heuristic == "Manhattan":
          h_score = manhattan(get_coordinates_of_all(node).values(), (get_coordinates_of_all(goal_state).values()))
        elif heuristic == "Hamming":
          h_score = hamming(get_coordinates_of_all(node).values(), (get_coordinates_of_all(goal_state).values()))
        new_h_score = h_score
        if flag == 0:
          print(f'h_score = {h_score}')
          print(node)
        node = tuple(map(tuple, node))
        priority_queue[node] = new_h_score 
  return correct_path
  

best_first(hard_example, goal_state, 0, "Manhattan")

```

```python
initial_list = np.array([[1,8,2], ["_",4,3], [7,6,5]])
correct_path = best_first(initial_list, goal_state, 1, "Hamming")
def best_first_path(correct_path):
  count = 0
  for array in correct_path:
    if count >=1:
      print(f'Move number: {count}')
    else:
      print("Initial State")
    print_board(array)
    count += 1
print("\n")
print("Correct Path ")
print("\n")
best_first_path(correct_path)
```
