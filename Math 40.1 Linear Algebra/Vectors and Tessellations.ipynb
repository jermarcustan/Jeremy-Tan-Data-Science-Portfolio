### Pythagorean Tiling

<div style = "text-align:center"><img src = "Pythagorean Tiling.svg" width = 10000>

```python
from sage.plot.colors import *

#vertices of big squares, small squares
bigVerts = list(map(vector, [[0, 0], [0, 12], [12, 12], [12, 0]]))
smallVerts = list(map(vector, [[12, 0], [12, 5], [17, 5], [17, 0]]))

#translation vectors
v1 = vector([12, 5])
v2 = vector([-5, 12])

n = 6

#big squares
bigSquares = sum([polygon([i*v1 + j*v2 + vert for vert in bigVerts], color = white, edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])

#small squares
smallSquares = sum([polygon([i*v1 + j*v2 + vert for vert in smallVerts], color = white, edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])

#signature
nameTxt = text('Loyola', (43, -49.5), fontsize = 13, fontweight = 'regular', color = black, rotation = 0)

tiling = bigSquares + smallSquares + nameTxt
tiling.show(xmin = -50, xmax = 50, ymin = -50, ymax = 50, figsize = 10, aspect_ratio = 1, axes = false)
tiling.save('pythagorean_bnw.svg', axes = false, aspect_ratio = 1, figsize = 15)
```

```python
from sage.plot.colors import *

#vertices of big purple square, big yellow square, vertices of red small squares

bigPurpleVerts = list(map(vector, [[0, 0], [0, 12], [12, 12], [12, 0]]))
bigYellowVerts = list(map(vector, [[12, 5], [12, 17], [24, 17], [24, 5]]))
redSmallVerts = list(map(vector, [[12, 0], [12, 5], [17, 5], [17, 0]]))

#translation vectors
v1 = vector([12, 5])
v2 = vector([-5, 12])
v3 = vector([7, 17])
v4 = vector([-17, 7])

n = 6

#big purple squares
bigPurpleSquares = sum([polygon([i*v3 + j*v4 + vert for vert in bigPurpleVerts], color = purple.lighter(), edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])

#big yellow squares
bigYellowSquares = sum([polygon([i*v3 + j*v4 + vert for vert in bigYellowVerts], color = yellow.lighter(0.5), edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])

#small red squares
smallRedSquares = sum([polygon([i*v1 + j*v2 + vert for vert in redSmallVerts], color = red.darker(), edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])

#signature
nameTxt = text( 'Loyola', (43, -49.5), fontsize = 13, fontweight = 'bold', color = white, rotation = 0 )

tiling = bigPurpleSquares + bigYellowSquares + smallRedSquares + nameTxt
tiling.show(xmin = -50, xmax = 50, ymin = -50, ymax = 50, figsize = 10, aspect_ratio = 1, axes = false)
tiling.save('pythagorean_colored.svg', axes = false, aspect_ratio = 1, figsize = 15)
```
