#### Transformed Figure

```python
# Codes for figure transformation (required to be used)

def rot(t): # counterclockwise rotation
    return matrix(RR, [[cos(t), -sin(t)], [sin(t), cos(t)]])

def rot_clockwise(t): # clockwise rotation
    return matrix(RR, [[cos(t), sin(t)], [-sin(t), cos(t)]])

def transFigure(motion, figure):
    return sum([circle((motion*matrix(cenCol[0]).transpose()).list(), 0.56, thickness = 0, fill = true, facecolor = cenCol[1]) for cenCol in figure])
```

```python
transFigure(rot(pi/4), squirtle).show(xmin = -20, xmax = 20, ymin = -20, ymax = 20, aspect_ratio = 1, figsize = 10, axes = true)
```

```python
transFigure(rot_clockwise(pi/4), Ivysaur).show(xmin = -100, xmax = 100, ymin = -100, ymax = 60, aspect_ratio = 1, figsize = 10, axes = true)
```

#### Animated Figure

```python
# Creates an animation by showing graphics frames sequentially

# Squirtle is moving counterclockwise while Ivysaur is moving clockwise

SquirtleFrames = [transFigure(rot(t), squirtle) for t in sxrange (0, 2*pi, 2*pi/20)] # list of frames
IvysaurFrames = [transFigure(rot_clockwise(t), Ivysaur) for t in sxrange (0, 2*pi, 2*pi/20)] # list of frames
RotatingSquirtle = animate(SquirtleFrames, xmin = -100, xmax = 100, ymin = -100, ymax = 100, figsize = 10, axes = false)
RotatingIvysaur = animate(IvysaurFrames, xmin = -100, xmax = 100, ymin = -100, ymax = 100, figsize = 10, axes = false)
RotatingPoke = RotatingSquirtle + RotatingIvysaur 
# I added the two animations together after the animate() so that the gif contains both animations with corresponding frames.
RotatingPoke.show()
RotatingPoke.save('animatedCharacters.gif', delay = 30, iterations = 0) # save as gif file
```
