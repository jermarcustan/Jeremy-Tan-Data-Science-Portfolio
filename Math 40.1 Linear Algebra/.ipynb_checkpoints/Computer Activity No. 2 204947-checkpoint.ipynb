```python
# Preliminary Computations

# Finding the formula for the vertices of the shapes

# The order of the vertices is important to make a proper shape. 
'''
We have two sets of hexagons: one for each color 

Each interior angle of the hexagon is 120 degrees.
A hexagon has 6 vertices.

Using trigonometry and the cosine and sine functions, we can calculate the coordinates of the 6 vertices in a hexagon.  
Angle used is 120 degrees divided by 2 = 60 degrees

Let length of hexagon: a 
Leftmost Vertex has coordinate (0,0)
Upperleft Vertex has coordinate (acos60, asin60)
Upperright Vertex has coordinate (a + acos60, asin60)
Rightmost Vertex has coordinate (2a,0)
Lowerright Vertex has coordinate (a + acos60, -asin60)
Lowerleft Vertex has coordinate (acos60, -asin60)
'''

'''
1st set of hexagons
The left vertex has coordinates (0,0)
'''
# Computing the vertices of the hexagon

# Let length of side, a = 6
# Convert degrees to radians

Hex_A = [0,0]
Hex_B = [6*cos(pi/3), 6*sin(pi/3)]
Hex_C = [6 + 6*cos(pi/3), 6*sin(pi/3)]
Hex_D = [12, 0]
Hex_E = [6 + 6*cos(pi/3), -6*sin(pi/3)]
Hex_F = [6*cos(pi/3), -6*sin(pi/3)]
'''
2nd set of hexagons
The hexagon directly on top of the first hexagon.
The vertices have coordinates of the first hexagon shifted vertically up by 2a*sin(pi/3)
Add 2a*sin(pi/3) to the y coordinate
'''
Hex_G = [0,12*sin(pi/3)]
Hex_H = [6*cos(pi/3), 6*sin(pi/3)+12*sin(pi/3)]
Hex_I = [6 + 6*cos(pi/3), 6*sin(pi/3) + 12*sin(pi/3)]
Hex_J = [12, 12*sin(pi/3)]
Hex_K = [6 + 6*cos(pi/3), -6*sin(pi/3) + 12*sin(pi/3)]
Hex_L = [6*cos(pi/3), -6*sin(pi/3) + 12*sin(pi/3)]
```

```python
# Preliminary Computations

'''
We have two sets of 3 triangles: upright and inverted triangles
'''

# Upright Triangles

'''
Each Equilateral Triangle has 3 vertices
Left Base has coordinate (0,0)
Right Base has coordinate (a,0)
Top Vertex has coordinate (a*cos(pi/3), a*sin(pi/3))
'''

# There are 3 upright triangles each with 3 vertices
# We have 9 vertices in total
# The coordinates are shifted up by double the height of the hexagon (height = 2a*sin(pi/3)) 
# Add 4a*sin(pi/3) to all of the y coordinates

Upr_A = [0, 24*sin(pi/3)]
Upr_B = [6*cos(pi/3), 6*sin(pi/3) + 24*sin(pi/3)]
Upr_C = [6, 24*sin(pi/3)]
Upr_D = [6 + 6*cos(pi/3), 6*sin(pi/3) + 24*sin(pi/3)]
Upr_E = [12, 24*sin(pi/3)]
Upr_F = [6, 24*sin(pi/3)]
Upr_G = [6 + 6*cos(pi/3), -6*sin(pi/3) + 24*sin(pi/3)]
Upr_H = [6*cos(pi/3), -6*sin(pi/3) + 24*sin(pi/3)]
Upr_I = [6, 24*sin(pi/3)]



# Inverted Triangles
# We still have 9 vertices, 3 vertices for each of the 3 equilateral triangles
# We likewise add the same shift in y (4a*sin(pi/3)) for the inverted triangles. 

Inv_A = [0, 24*sin(pi/3)]
Inv_B = [6*cos(pi/3), -6*sin(pi/3) + 24*sin(pi/3)]  
Inv_C = [6, 24*sin(pi/3)]
Inv_D = [6*cos(pi/3), 6*sin(pi/3) + 24*sin(pi/3)]
Inv_E = [6 + 6*cos(pi/3), 6*sin(pi/3) + 24*sin(pi/3)]
Inv_F = [6, 24*sin(pi/3)]
Inv_G = [12, 24*sin(pi/3)]
Inv_H = [6 + 6*cos(pi/3), -6*sin(pi/3) + 24*sin(pi/3)]
Inv_I = [6, 24*sin(pi/3)]
```

```python
# Preliminary Computations

'''
I noticed that the hexagons are evenly spaced from each other in the diagonals. 
The hexagons on top of each other are also evenly spaced.
All of the shapes can be represented by the linear combination of two translation vectors. 

There are 2 translation vectors.
1.) vertical translation.
2.) diagonal translation. 
 
Get the vector between identical points in 2 consecutive shapes.
We use vector subtraction to get the translation vecotr. 
'''


# A.) Hexagon 1st set (purple)

# POINT: left vertex of hexagon

# Vertical Translation

vertical_hex1 = vector([0, 18*sqrt(3)]) - vector([0,0])

# Diagonal Translation 

diagonal_hex1 = vector([9, 9*sqrt(3)]) - vector([0,0])

# B.) Hexagon 2nd set (yellow)

# POINT: left vertex of hexagon

# Vertical Translation

vertical_hex2 = vector([0, 24*sqrt(3)]) - vector([0,6*sqrt(3)])

# Diagonal Translation 

diagonal_hex2 = vector([9, 15*sqrt(3)]) - vector([0,6*sqrt(3)])

# C.) Upright Triangles (purple)

# POINT: intersection of the three upright triangles

# Vertical Translation

vertical_ut = vector([6, 30*sqrt(3)]) - vector([6, 12*sqrt(3)])

# Diagonal Translation 

diagonal_ut = vector([15, 21*sqrt(3)]) - vector([6, 12*sqrt(3)])

# D.) Inverted Triangles (yellow)

# POINT: intersection of the three inverted triangles

# Vertical Translation

vertical_it = vector([6, 30*sqrt(3)]) - vector([6, 12*sqrt(3)])

# Diagonal Translation 

diagonal_it = vector([15, 21*sqrt(3)]) - vector([6, 12*sqrt(3)])

print(f"{vertical_hex1} {diagonal_hex1}")
print(f"{vertical_hex2} {diagonal_hex2}")
print(f"{vertical_ut} {diagonal_ut}")
print(f"{vertical_it} {diagonal_it}")
print("The translation vectors for all of the shapes are the same.")
```

```python
# Black and White Tessellation
from sage.plot.colors import *

# Let length of one side of hexagon: 6
# Hexagon has 6 vertices

# 1st set of hexagons
HexagonVerts_set_1 = list(map(vector, [[0, 0], [3,3*sqrt(3)], [9, 3*sqrt(3)], [12,0], [9,-3*sqrt(3)], [3, -3*sqrt(3)]]))
# 2nd set of hexagons
HexagonVerts_set_2 = list(map(vector, [[0, 6*sqrt(3)], [3,9*sqrt(3)], [9, 9*sqrt(3)], [12,6*sqrt(3)], [9,3*sqrt(3)], [3, 3*sqrt(3)]]))
# Upright Triangles
TriangleVerts = list(map(vector, [[0, 12*sqrt(3)], [3,15*sqrt(3)], [6, 12*sqrt(3)], [9, 15*sqrt(3)], [12,12*sqrt(3)], [6, 12*sqrt(3)], [9,9*sqrt(3)], [3, 9*sqrt(3)], [6, 12*sqrt(3)]]))
# Note: I did not need to put the coordinates of the three inverted triangles in the black and white figure. I only needed it for the colored figure. 


# Translation vectors
v1 = vector([9, 9*sqrt(3)]) # diagonal translation
v2 = vector([0, 18*sqrt(3)]) # vertical translation

n = 20


Hexagon_1 = sum([polygon([i*v1 + j*v2 + vert for vert in HexagonVerts_set_1], color = white, edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])
Hexagon_2 = sum([polygon([i*v1 + j*v2 + vert for vert in HexagonVerts_set_2], color = white, edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])
Triangle = sum([polygon([i*v1 + j*v2 + vert for vert in TriangleVerts], color = white, edgecolor = black, thickness = 2) for i in [-n..n] for j in [-n..n]])



#signature
nameTxt = text('Jeremy', (43, -49.5), fontsize = 13, fontweight = 'regular', color = black, rotation = 0)

tiling = Hexagon_1 + Hexagon_2 + Triangle + nameTxt
tiling.show(xmin = -50, xmax = 50, ymin = -50, ymax = 50, figsize = 10, aspect_ratio = 1, axes = false)
tiling.save('black-and-white.svg', axes = false, aspect_ratio = 1, figsize = 15)
```
