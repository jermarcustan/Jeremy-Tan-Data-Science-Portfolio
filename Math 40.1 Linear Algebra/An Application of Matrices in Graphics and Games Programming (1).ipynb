#### Transformed Figure

```python
# Rotation about the origin by an angle t (counterclockwise if positive)

def rot(t):
    return matrix(RR, [[cos(t), -sin(t)], [sin(t), cos(t)]])
```

```python
# Transforms a figure consisting of pairs [point, color] using motion

def transFigure(motion, figure):
    return sum([circle((motion*matrix(cenCol[0]).transpose()).list(), 0.56, thickness = 0, fill = true, facecolor = cenCol[1]) for cenCol in figure])

transFigure(rot(pi/4), mario).show(xmin = -20, xmax = 20, ymin = -20, ymax = 20, aspect_ratio = 1, figsize = 10, axes = true)
```

#### Animated Figure

```python
# Creates an animation by showing graphics frames sequentially

marioFrames = [transFigure(rot(t), mario) for t in sxrange (0, 2*pi, 2*pi/20)] # list of frames

rotatingMario = animate(marioFrames, xmin = -20, xmax = 20, ymin = -20, ymax = 20, figsize = 10, axes = true)
rotatingMario.show()
rotatingMario.save('mario.gif', delay = 30, iterations = 0) # save as gif file
```
